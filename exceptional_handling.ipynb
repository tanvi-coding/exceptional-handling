1.What is the difference between interpreted and compiled languages?

  ->Interpreted languages execute code line by line at runtime (e.g., Python).
  ->Compiled languages convert the entire code into machine code before execution (e.g., C, C++).

2.What is exception handling in Python?

  ->Exception handling is a mechanism to handle runtime errors using try, except, else, and finally blocks so that the program does not crash unexpectedly.

3.What is the purpose of the finally block in exception handling?

   ->The finally block always executes, whether an exception occurs or not. It is mainly used for cleanup actions like closing files or releasing resources.

4.What is logging in Python?

  ->Logging is used to record program events, errors, and messages during       execution for debugging and monitoring purposes.

5.What is the significance of the __del__ method in Python?

   ->The __del__ method is a destructor that is called when an object is about to be destroyed. It helps release resources like memory or file handles.

6.What is the difference between import and from ... import in Python?

   ->import module imports the entire module.

   ->from module import name imports specific items from a module.

7.How can you handle multiple exceptions in Python?

8.What is the purpose of the with statement when handling files in python?
